BASIC INFO ABOUT THE DATAFRAME

```python
print(dogs.head())  ###shows First 5 rows...similarly .tail() will show the last 5
print( "MEAN_HEIGHT =",dogs["height_cm"].mean()  )  ##similarly .max .min .median .mode .std .var ....
```

```python
""" .agg method and PERCENTILE ( .quantile() )"""
def pct30(column):
    return column.quantile(0.3)
def pct40(column):
    return column.quantile(0.4)
print(   dogs[["weight_kg","height_cm"]].agg(pct30)   )  #### returns the 30th percentile of height and weight 
print("\n######\n")

print(  dogs["weight_kg"].agg([pct30 ,pct40])  ) ### returns 
```

```python
""" cumulative .cumsum() , .cummin(), cummax()... """
print( dogs[["weight_kg"]].cumsum() )
print( dogs[["weight_kg"]].cummin())
```

GROUPBY

```python
""" NOT OPTIMIZED METHOD """
print (  dogs[ dogs["color"] == 'Black' ]["weight_kg"].mean()  )
print (  dogs[ dogs["color"] == 'Brown' ]["weight_kg"].mean()  )
print (  dogs[ dogs["color"] == 'White' ]["weight_kg"].mean()  )
print (  dogs[ dogs["color"] == 'Gray' ]["weight_kg"].mean()  )
print (  dogs[ dogs["color"] == 'Tan' ]["weight_kg"].mean()  )
```

```python
""" OPTIMIZED METHOD """
dogs.groupby("color")[["weight_kg"]].mean()
```

```python
"""   multiple column grouping and .agg   """

dogs.groupby(["color","breed"])[["weight_kg","height_cm"]].agg([min,max,sum])
```

### PIVOT TABLE ###

```python
"""  showing info about a particular paramter (eg: weight_kg in this case, for all colors and breeds)   """

"""
NaN is showing as there's no value present, there are no Black Brown Gray or White Chihuahua for example 

"""
dogs.pivot_table(values = "weight_kg", index = 'color', columns = 'breed')
```

```python
""" to replace the NaN we use fill_value"""
"""
margins = True  gives the mean value for each row and columns...
"""
dogs.pivot_table(values = "weight_kg", index = 'color', columns = 'breed', fill_value = 0, margins = True)
```
