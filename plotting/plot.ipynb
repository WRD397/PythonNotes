### Plotting TIME SERIES & GRID PLOT

```python
climate_change = pd.read_csv("E:\VSCode\Python\climate_change.csv")
climate_change.head()
```

```python
climate = pd.read_csv("E:\VSCode\Python\climate_change.csv", parse_dates = ['date'], index_col = 'date')
climate.head()
```

```python
fig, ax = plt.subplots(2,1,figsize = (16,8), sharex = True)
ax[0].plot(climate.index, climate.co2, linestyle = '-.', color='green')
ax[1].plot(climate.index, climate.relative_temp, linestyle = '-',color = 'red')
ax[0].set_ylabel('co2(ppm)')
ax[1].set_ylabel('relative_temp')
ax[1].set_xlabel('Year')

ax[0].grid()
ax[1].grid()
plt.show()
```

### PLOTTING ON SAME PLOT                                                                                                 
## ax.twinx()

```python
fig, ax = plt.subplots()
ax.plot(climate.index, climate.co2, color='green')
ax.plot(climate.index, climate.relative_temp, color = 'red')
ax.set_ylabel('co2(ppm) / relative_temp')
ax.set_xlabel('Year')
plt.show()
```

...thus due to scale mismatch plots are not that much useful...to handle this we will apply the below method

```python
fig, ax = plt.subplots(figsize=(16,8))

ax.plot(climate.index, climate.co2, color='red')
ax.set_xlabel("Time")
ax.set_ylabel('co2(ppm)',color='red')
#or, ax.yaxis.label.set_color('green')
ax.tick_params('y',colors='red')

ax2 = ax.twinx()

ax2.plot(climate.index,climate.relative_temp, color='blue')
ax2.set_xlabel("Time")
ax2.set_ylabel('relative_temp',color='blue')
#or,  ax2.yaxis.label.set_color('blue')
ax2.tick_params('y',colors='blue')

ax.grid(color='red')
ax2.grid(color='blue')

plt.show()
```

#### implementing a FUNCTION containing these code lines...

```python
def plot_timeseries(axes, x, y, color, ylabel):
    axes.plot(x,y,color)
    axes.set_xlabel('Time')
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y',colors=color)
    axes.grid(color=color)
```

```python
fig,ax = plt.subplots(figsize=(16,8))
plot_timeseries(ax, climate_change.index, climate_change.co2, color='red', ylabel='co2(ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change.relative_temp, color='blue', ylabel='Relative_temp')
plt.show()
```

### HISTOGRAM

```python
summer = pd.read_csv("E:\VSCode\Python\summer2016.csv",index_col = [0])
summer.head()
```

```python
mens_rowing = summer[(summer['Sex']=='M') & (summer['Sport']=='Rowing')]
mens_gymnastic = summer[(summer['Sex']=='M') & (summer['Sport']=='Gymnastics')]
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label='Rowing',bins=5)
ax.hist(mens_gymnastic["Height"], label='Height',bins=5)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label='Rowing',bins=5, histtype = 'step')
ax.hist(mens_gymnastic["Height"], label='Height',bins=5, histtype = 'step')
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

### ERROR BARS

```python
fig, ax = plt.subplots()
ax.bar("Rowing",mens_rowing['Height'].mean(),yerr = mens_rowing['Height'].std())
ax.bar("Gymnastics", mens_gymnastic['Height'].mean(), yerr = mens_gymnastic['Height'].std())
ax.set_ylabel('Height(cm)')
plt.show()
```

...so here we are getting an idea about also deviation ie. an extra information apart from the mean of the whole distribution in a single histogram.
