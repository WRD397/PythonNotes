```python
df2 = pd.read_csv("E:\VSCode\PythonNotes\seaborn_notes\student-alcohol-consumption.csv")
hue_colors = {'Rural' : 'black',
                'Urban' : 'red'}
sns.scatterplot(x="absences",y='G3', data=df2, hue='location', hue_order=['Rural','Urban'], palette = hue_colors)                
```

```python
palette_colors = {'Rural': "black", 'Urban': "yellow"}
#Creating a count plot of school with location subgroups
sns.countplot(x='school',data=df2, hue="location", palette=palette_colors )
```
